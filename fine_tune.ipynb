```python
dataset = Dataset.from_list(data)

model_name = "Qwen/Qwen1.5-0.5B-Chat"

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen1.5-0.5B-Chat")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen1.5-0.5B-Chat")
```

This function tokenizes each dialogue example by converting the text into input IDs, with padding and truncation up to 64 tokens.
It also sets the labels as a copy of the input IDs, which is needed for training language models to predict the next word.

```python
def tokenize_function(examples):
    tokens = tokenizer(examples["text"],
                       truncation=True,
                       max_length=64,
                       padding="max_length")
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens
```

This line applies the tokenize_function to the entire dataset using .map(), which processes all examples in batches.
The result is a new dataset with input IDs and labels, ready for training the model.



```python
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

This sets up training configurations like output folder, batch size, number of epochs, logging frequency, and saving checkpoints.
no_cuda=True forces training on CPU (disable GPU), and fp16=False means no mixed precision — suitable if you don’t have a GPU.



```python
training_args = TrainingArguments(
    output_dir="./fast_finetune",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=50,
    save_total_limit=2,
    no_cuda=True,
    fp16=False
)
```

This creates a Trainer object that connects the model, training settings, and the tokenized dataset for training.
It handles the entire training loop automatically, including saving and logging.



```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)
```

This starts the training process using the settings and data you provided, fine-tuning the model on your dataset.
It runs through the data for the specified number of epochs and updates the model weights.

```python
trainer.train()
```

These commands save your fine-tuned model and tokenizer files into the folder ./fast_finetuned so you can load and use them later.

```python
trainer.save_model('./fast_finetuned')
tokenizer.save_pretrained('./fast_finetuned')
```

This function takes a text prompt, tokenizes it, and generates a response from the fine-tuned model up to max_length tokens.
It then decodes the generated tokens back to readable text and returns the chatbot’s reply.

```python
def generate_response(prompt, max_length = 64):
    inputs = tokenizer(prompt, return_tensors='pt')
    outputs = model.generate(**inputs, max_length = max_length)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
prompt= 'What is the quantity of Copy?\n'
print(("chatbot response:",generate_response(prompt) ))
```
